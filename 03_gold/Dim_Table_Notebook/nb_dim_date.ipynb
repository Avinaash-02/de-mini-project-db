```python
%sql
CREATE OR REPLACE TABLE de_mini_project.03_gold_schema.dim_date AS
SELECT DISTINCT
    CAST(order_purchase_timestamp AS DATE) AS date,
    YEAR(order_purchase_timestamp) AS year,
    MONTH(order_purchase_timestamp) AS month
FROM de_mini_project.02_silver_schema.orders;
```
