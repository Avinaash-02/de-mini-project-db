```python
%sql
CREATE OR REPLACE TABLE de_mini_project.03_gold_schema.cube_revenue AS
SELECT
    year,
    month,
    product_category_name,
    state,

    ROUND(SUM(revenue),2) AS total_revenue,
    ROUND(COUNT(DISTINCT order_id),2) AS total_orders,
    ROUND(AVG(delivery_days),2) AS avg_delivery_days

FROM de_mini_project.03_gold_schema.curated_orders

GROUP BY CUBE (
    year,
    month,
    product_category_name,
    state
);
```
