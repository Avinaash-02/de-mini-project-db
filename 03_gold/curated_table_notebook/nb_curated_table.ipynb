```python
%sql
CREATE OR REPLACE TABLE de_mini_project.03_gold_schema.curated_orders AS
SELECT
    o.order_id,
    o.customer_id,
    o.order_date,
    d.year,
    d.month,

    c.unique_id,
    g.city,
    g.state,

    oi.product_id,
    p.product_category_name,

    oi.seller_id,
    ROUND(oi.price + oi.freight_value,2) AS revenue,

    o.delivery_days,
    o.is_late_delivery,
    o.order_status,

    pay.payment_type,
    r.review_score

FROM de_mini_project.03_gold_schema.fact_orders o
JOIN de_mini_project.03_gold_schema.fact_order_items oi 
    ON o.order_id = oi.order_id
JOIN de_mini_project.03_gold_schema.dim_date d 
    ON o.order_date = d.date
JOIN de_mini_project.03_gold_schema.dim_customers c 
    ON o.customer_id = c.customer_id
LEFT JOIN de_mini_project.03_gold_schema.dim_geolocation g 
    ON c.zip_code_prefix = g.zip_code_prefix
JOIN de_mini_project.03_gold_schema.dim_products p 
    ON oi.product_id = p.product_id
JOIN de_mini_project.03_gold_schema.dim_sellers s 
    ON oi.seller_id = s.seller_id
LEFT JOIN de_mini_project.03_gold_schema.fact_payments pay 
    ON o.order_id = pay.order_id
LEFT JOIN de_mini_project.03_gold_schema.fact_reviews r 
    ON o.order_id = r.order_id;
```
