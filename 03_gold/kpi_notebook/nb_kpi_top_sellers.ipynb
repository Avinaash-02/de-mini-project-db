```python
%sql
CREATE OR REPLACE TABLE de_mini_project.03_gold_schema.kpi_top_sellers AS
SELECT seller_id, ROUND(SUM(revenue),2) revenue
FROM de_mini_project.03_gold_schema.curated_orders
GROUP BY seller_id
ORDER BY revenue DESC LIMIT 10;
```
