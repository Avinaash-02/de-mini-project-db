```python
%sql
-- 12 -kpi revenue by seller
CREATE OR REPLACE TABLE de_mini_project.03_gold_schema.kpi_revenue_by_sellers AS
SELECT seller_id , ROUND(SUM(revenue),2) AS total_revenue
FROM de_mini_project.03_gold_schema.curated_orders
GROUP BY seller_id 
ORDER BY total_revenue DESC
LIMIT 10;
```
