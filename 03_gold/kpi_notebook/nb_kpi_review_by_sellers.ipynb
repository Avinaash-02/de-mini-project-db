```python
%sql
CREATE OR REPLACE TABLE de_mini_project.03_gold_schema.kpi_reviews_seller AS
SELECT seller_id, ROUND(AVG(review_score),2) as avg_rating
FROM de_mini_project.03_gold_schema.curated_orders
GROUP BY seller_id;
```
