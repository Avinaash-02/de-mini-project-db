```python
%sql
CREATE OR REPLACE TABLE de_mini_project.03_gold_schema.kpi_revenue_monthly AS
SELECT year, month, ROUND(SUM(revenue),2) AS total_revenue
FROM de_mini_project.03_gold_schema.curated_orders
GROUP BY year, month;
```
